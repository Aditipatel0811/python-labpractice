```python
from tkinter import * 
base = Tk()  
base.geometry("550x550")  
base.title('Airline booking')

lbl_0 = Label(base, text="Airline booking")   
lbl_0.pack()  
  

lbl_1 =Label(base, text= "enter your name")  
lbl_1.pack()  
  
en_1 = Entry(base)  
en_1.pack() 
   
lbl_2 = Label(base, text="enter your registered mail")  
lbl_2.pack()  
  
en_2 = Entry(base)  
en_2.pack() 

lbl_3 = Label(base, text="enter your departure")  
lbl_3.pack() 
  
en_3 = Entry(base)  
en_3.pack() 

lbl_4= Label(base, text="enter your Arrival")  
lbl_4.pack()  
  
en_4 = Entry(base)  
en_4.pack() 
   
lbl_5 = Label(base, text="Gender")  
lbl_5.pack()   
     
vars = IntVar()  
   
Radiobutton(base, text="Male", padx= 5, variable= vars, value=1).pack()   
Radiobutton(base, text="Female", padx= 20, variable= vars, value=2).pack()  
  
  
lbl_6 = Label(base, text="enter your Age")  
lbl_6.pack() 
  
en_6 = Entry(base)  
en_6.pack() 

lbl_7=Label(base, text ="Country")  
lbl_7.pack() 
  
 
list_of_cntry=[ 'India' ,'Canada' , 'US' ,'Germany' ,'UK']  
    
cv = StringVar()  
drplist = OptionMenu(base, cv, *list_of_cntry)  
drplist.config(width=15)  
cv.set('Select your Country')  
drplist.pack()
    
Button(base, text='confirm' , width=20, bg="brown",fg='white').pack() 
  
base.mainloop()  

print("Airline booking form is created susccesfully..........")
```
